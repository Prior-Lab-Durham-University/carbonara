```python
# name the PDB it needs to be in the same folder as this notebook
pdb_file = 'molecules/heat_shock/model_relaxed1.pdb'

# load in the saxs file
SAXS_file = 'molecules/heat_shock/conc1.dat'

# Give the molecule a name, it doesn't matter what, just make it memoreable
mol_name = 'shock_protein'

# Write the moelcule directory
working_path = CDT.setup_working_directory(mol_name)
```

```python
# need to merge A,B C,D ... etc.. -> dimers become 'monomers'

# Matteo Degiacomi's BioBox to manipulate PDB files
M = CDT.pdb_2_biobox(pdb_file)

M.data['chain'] = M.data['chain'].map({'A':'A', 'B':'A', 'C':'B', 'D':'B', 'E':'C', 'F':'C', 'G':'D', 'H':'D', 'I':'E', 'J':'E', 'K':'F', 'L':'F'})
```

```python
M.data
```

```python
coords = CDT.extract_CA_coordinates(M)
sequence = CDT.extract_sequence(M)
sequence[sequence.astype(str)=='nan'] = 'G'

# check if this is a single file multi domain protein: look for big gaps in calphas
coordssplit,sequencesplit,splitList = CDT.splitMoleculeNoSec(coords,sequence)

coordssplit_new = []
sequencesplit_new = []
splitList_new = splitList[1::2]

# merge adjacent lists: 12 -> 6
for i in range(6):
    coordssplit_new.append(np.concatenate( [coordssplit[2*i], coordssplit[(2*i)+1]]))
    sequencesplit_new.append(np.concatenate( [sequencesplit[2*i], sequencesplit[(2*i)+1]]))

# use kappa tau method to identify secondary structure
# secondarysplit = [CDT.findSimilar(CDT.getKapTauList(coordssplit_new[i]),0.2) for i in range(len(coordssplit_new))]

# secondary = CDT.DSSP_structure_extractor('molecules/heat_shock/model_Aonly.pdb')
ss = ['-']*28 + ['S']*116 + ['-']*6 + ['S']*8  # 4 sections per monomer - 8 per dimer
ss = np.asarray( 2*ss )
secondarysplit = [ss, ss, ss, ss, ss, ss]
```

```python
# Writes all the coords, fingerprint, mixture & SAXS in Carbonara's expected input / file struct
CDT.write_fingerprint_file(len(coordssplit_new), sequencesplit_new, secondarysplit, working_path)    
CDT.write_coordinates_file(coords,working_path)
CDT.write_mixture_file(working_path)
CDT.write_saxs(SAXS_file, working_path)
```

```python
# adding restraints

reach_around_pairs = [ ['A', 'D'], ['B', 'H'], ['C', 'L'], ['D', 'J'], ['E', 'C'], ['F', 'G'],
                       ['G', 'I'], ['H', 'K'], ['I', 'F'], ['J', 'A'], ['K', 'B'], ['L', 'E']]

p_map = {'A':1, 'B':2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7, 'H':8, 'I':9, 'J':10, 'K':11, 'L':12}

reach_around_pairs_new = []

for pair in reach_around_pairs:
    new_pair = []
    for p in pair:
        new_pair.append(p_map[p])
    reach_around_pairs_new.append(new_pair)

# add dimer normalised contact restraints
dimer_len = 158

restraint_pairs = [[154, 129], [156, 131]]

restraint_pairs_new = []
fixed_distances = []

for pair in reach_around_pairs_new:

    # print(pair[0], pair[1])
    restraint_pairs_new.append( [ int(154 + (pair[0]-1)*dimer_len), int(129 + (pair[0]-1)*dimer_len) ] )
    fixed_distances.append(6.1)

    restraint_pairs_new.append( [ int(156 + (pair[0]-1)*dimer_len), int(131 + (pair[0]-1)*dimer_len) ] )
    fixed_distances.append(6.0)
    
```

```python
CDT.translate_distance_constraints(restraint_pairs_new, coords, splitList_new, working_path, fixed_distances)
```
